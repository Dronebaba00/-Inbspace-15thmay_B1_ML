```python
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from datetime import date, datetime
import numpy as np
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_train.head(5)
```

```python
#Creating a flag for each type of restaurant
df_train['Type_IL'] = np.where(df_train['Type'] == 'IL', 1, 0)
df_train['Type_FC'] = np.where(df_train['Type'] == 'FC', 1, 0)
df_train['Type_DT'] = np.where(df_train['Type'] == 'DT', 1, 0)
```

```python
#Creating a flag for 'Big Cities'
df_train['Big_Cities'] = np.where(df_train['City Group'] == 'Big Cities', 1, 0)

#Converting Open_Date into day count
#Considering the same date the dataset was made available
df_train['Days_Open'] = (pd.to_datetime('2021-06-13') - pd.to_datetime(df_train['Open Date'])).dt.days
```

```python
#Removing unused columns
df_train = df_train.drop('Type', axis=1)
df_train = df_train.drop('City Group', axis=1)
df_train = df_train.drop('City', axis=1)
df_train = df_train.drop('Open Date', axis=1)
```

<b> Test Data </b>

```python
df_test.head(5)
```

```python
#Adjusting test data as well
df_test['Type_IL'] = np.where(df_test['Type'] == 'IL', 1, 0)
df_test['Type_FC'] = np.where(df_test['Type'] == 'FC', 1, 0)
df_test['Type_DT'] = np.where(df_test['Type'] == 'DT', 1, 0)
df_test['Big_Cities'] = np.where(df_test['City Group'] == 'Big Cities', 1, 0)
df_test['Days_Open'] = (pd.to_datetime('2021-06-13') - pd.to_datetime(df_test['Open Date'])).dt.days
df_test = df_test.drop('Type', axis=1)
df_test = df_test.drop('City Group', axis=1)
df_test = df_test.drop('City', axis=1)
df_test = df_test.drop('Open Date', axis=1)
```

```python
df_train.dtypes
```

<b> Splitting data into X and Y </b>

```python
from sklearn import model_selection
from sklearn import linear_model
X = df_train.drop(['Id', 'revenue'], axis=1)
Y = df_train.revenue
```

```python
model = RandomForestRegressor(n_estimators=150)
model.fit(X, Y)
test_predicted = pd.DataFrame()
test_predicted['Id'] = df_test.Id
test_predicted['Prediction'] = model.predict(df_test.drop('Id', axis=1))
test_predicted.to_csv('submission-random-forest.csv', index=False)
test_predicted.head()
score=model.score(X,Y)
print("Accuracy of model is :",score*100 ,"%")
```
